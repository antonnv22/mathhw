```python
import numpy as np
import matplotlib.pyplot as plt

# Define the derivative function
def f_prime(x):
    return 2 * x * np.cos(x) / np.sin(x) - x**2 / np.sin(x)**2

# Plot the derivative function
x_values = np.linspace(-10, 10, 1000)
# Avoid points where sin(x) = 0
x_values = x_values[np.abs(np.sin(x_values)) > 0.01]
y_values = f_prime(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r"$f'(x) = 2x \cdot \cot(x) - x^2 \cdot \csc^2(x)$")
plt.title("График производной функции")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-500000, 10)
plt.grid(True)
plt.legend()
plt.show()
```
